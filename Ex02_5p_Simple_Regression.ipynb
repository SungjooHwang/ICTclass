```python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model
plt.rcParams['figure.figsize'] = (12.0, 9.0)


# Preprocessing Input data
data = pd.read_csv('Ex05_3_Data.csv')  #Dataframe
X = data[['X']]
Y = data[['Y']]


plt.scatter(X, Y)
plt.show()
```

```python
# Building the model
model = sklearn.linear_model.LinearRegression()

# Training the model
model.fit(X,Y)
model.intercept_, model.coef_
```

```python
# Making predictions
t0, t1 = model.intercept_[0], model.coef_[0]
Y_pred = t0 + t1 * X 
plt.scatter(X, Y)
plt.plot([np.min(X), np.max(X)], [np.min(Y_pred), np.max(Y_pred)], color='red') # predicted
plt.show()
```
